```julia
using Plots
t1 = -5:0.01:5
```

```julia
signal1 = 5 * sin.(2 * π .* t1) .* cos.(π .* t1 .- 8)
```

```julia
plot(t1, signal1, xlabel = "Time (s)", ylabel = "Amplitude",title = "Signal 1", show = true)
```

```julia
t= -10:0.01:10
signal2 = 5 * exp.(-0.2 .*t) .* sin.(2*π .*t)

plot(t, signal2, xlabel = "Time(s)", ylabel = "Amplitude", title = "Signal 2")
```
